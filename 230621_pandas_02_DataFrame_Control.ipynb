# 데이터프레임 고급 인덱싱

* 인덱싱(indexing) : 데이터프레임에서 특정한 데이터만 골라내는 것
 * 라벨, 라벨 리스트, 인덱스데이터(정수) 슬라이스의 3가지 인덱싱 값을 사용하여 인덱싱
* 그런데 Pandas는 Numpy 행렬과 같이 쉼표를 사용한 (행 인덱스, 열 인덱스) 형식의 2차원 인덱싱을 지원하기 위해 다음과 같은 특별한 인덱서(indexer) 속성도 제공
* `loc` : 라벨값 기반의 2차원 인덱싱
* `iloc` : 순서를 나타내는 정수 기반의 2차원 인덱싱

## `loc` 인덱서
```
df.loc[행 인덱싱 값]
df.loc[행 인덱싱 값, 인덱싱 값]
```
* 행 인덱싱 값 : 정수 또는 행 인덱스 데이터
* 열 인덱싱 값 : 라벨 문자열
---
* 인덱스 데이터
* 인덱스 데이터 슬라이스
* 인덱스 데이터 리스트
* 같은 행 인덱스를 가지는 불리언 시리즈 (행 인덱싱의 경우)
* 위의 값들을 반환하는 함수

````python
import numpy as np
import pandas as pd
````

````python
df = pd.DataFrame(
    np.arange(10, 22).reshape(3, -1),
    columns = list('ABCD'),
    index = list('abc')
)
df
````

### 인덱싱 값을 하나만 받는 경우
* 만약 loc 인덱서를 사용하면서 인덱스를 하나만 넣으면 행(row)을 선택

````python
df.loc['a'], df.loc['b'], df.loc['c'] # 문자열 인덱스
````

````python
# 인덱스 데이터의 슬라이스도 가능
df.loc['b':'c']
````

````python
# 기본 인덱싱과 동일
df['b':'c']
````

````python
# 인덱스데이터의 리스트도 됨
df.loc[['b', 'c', 'c']]
````

````python
# 이 때는 loc를 쓰지 않으면 KeyError 오류가 발생
````

````python
# 데이터프레임와 같은 인덱스를 가지는 불리언 시리즈도 행을 선택하는 인덱싱값으로 쓸 수 있음
df.A > 15
````

````python
df.loc[df.A > 15] # loc -> 불리언 배열 인덱싱, 라벨 인덱싱 모두 성립한다.
````

````python
# 인덱스 대신 인덱스 값을 반환하는 함수를 사용할 수도 있음
def select_rows(df):
  return df.A > 15  # 벡터화 -> df와 같은 인덱스를 가지는 불리언 배열
````

````python
select_rows(df)
````

````python
df[select_rows(df)], df.loc[select_rows(df)]
````

````python
# loc 인덱서가 없는 경우에 사용했던 라벨 인덱싱이나 라벨 리스트 인덱싱은 불가능
# df.loc["A"] # loc에 column 키 값을 넣으면 -> 에러 # keyError : 'A'
````

````python
# 원래 (행) 인덱스값이 정수인 경우에는 슬라이싱도 라벨 슬라이싱 방식을 따르게 됨
# 즉, 슬라이스의 마지막 값이 포함
df2 = pd.DataFrame(np.arange(10, 26).reshape(4, -1),
                   columns = ['A', 'B', 'C', 'D'])
df2
````

````python
# df2[1:2] 와 df2.loc[1:2] 의 차이점

# .loc 는 [n:m] n부터 m 숫자의 포함 까지의 인덱스 값
````

````python
df2[1:2]
````

````python
df2.loc[1:2]
````

|인덱싱 값|가능|결과|자료형|추가사항|
|:-|:-|:-|:-|:-|
|행 인덱스 값( 정수)|O|행|시리즈|
|행 인덱스 값 (정수) 슬라이스|O|행|데이터프레임|`loc`가 없는 경우와 같음|
|행 인덱스 값 (정수) 리스트|O|행|데이터프레임|
|불리언 시리즈|O|행|데이터프레임|시리즈의 인덱스가 데이터프레임의 행 인덱스와 같아야 함|
|불리언 시리즈를 반환하는 함수|O|행|데이터프레임
|열 라벨|X|||loc가 없는 경우에만 쓸 수 있음|
|열 라벨 리스트|X|||loc가 없는 경우에만 쓸 수 있음|

## 정렬 (sort)

* `sort_values` : 데이터 값 기준으로 정렬.
> 오름차순 -> 행이 늘어나는 방향 -> 데이터 나열되는 방향으로 -> 데이터가 커지는 방향을 일치시켜주겠다 (ascending)<br>

## `apply` 변환

* sum, mean 이미 정의된 함수/메소드. 어떠한 작업을 해줄지 이미 정해져있음
* 행이나 열 단위로 복잡한 데이터 처리 -> `apply` 메소드 사용
* 인수로 행 또는 열을 받는 함수 (axis)를 `apply` 메소드의 인수로 넣으면 각 열(또는 행)을 반복하여 그 함수에 적용시킴


````python
df4 = pd.DataFrame({
    'A': [1, 3, 4, 3, 4],
    'B': [2, 3, 1, 2, 3],
    'C': [1, 5, 2, 4, 4],
})
df4
````

````python
# 각 열의 최댓값과 최솟값의 차이를 구하는 연산
df4.max() - df4.min()
````

````python
df4.columns
````

````python
df4.index
````

````python
# 람다 함수
lambda x: (x) # lambda 인자들 :인자를 사용해서 반환될 표현식
# 익명함수 -> lambda 이름을 줄 필요 X (def ... 변수에 할당) -> 사용되고 휘발
# x -> 처리할 데이터 -> apply 변수
# df3.apply(lambda x: x.max() - x.min())
df4.apply(lambda x: x.max() - x.min(), axis=0)
````

````python
# [열 기준으로 apply]
# 모든 행의 첫번째 글자의 대문자

titanic.sex.apply(lambda x : x[0].upper())
````

* `reset_index` : 기존의 행 인덱스를 제거하고, 해당 인덱스를 새로운 열로 추가
